## Make  dataframe

```python
import numpy as np
import pandas as pd

with open("data/lubm/lubm_2m","r") as f:
    full = f.readlines()
    
    sub = []
    pro = []
    ob = []
    for i in full:
        sub.append(i.split(" ")[0].strip())
        pro.append(i.split(" ")[1].strip())
        ob.append(i.split(" ")[2].strip())
    
raw_data = {"object" : ob,"property" : pro, "subject" : sub }
df = pd.DataFrame(raw_data, columns=['subject', 'property', 'object'])
df
```

## make  class of  list

```python
listOfClass = df[(df['property'] == "type")]['object'].drop_duplicates().values.tolist()
listOfClass
```

## make  property table

```python
df_university = df[df['object'] == listOfClass[1]][['subject']]
df_university
```

```python
result = pd.merge(df, df_university, on='subject')
listOfProperty = result['property'].drop_duplicates().values.tolist()
print(listOfProperty)
p1 = result[result['property'] == listOfProperty[0]][['subject', 'object']]
p2 = result[result['property'] == listOfProperty[1]][['subject', 'object']]
p3 = result[result['property'] == listOfProperty[2]][['subject', 'object']]
```

```python
result = pd.merge(p1, p2, on='subject')
result
```
